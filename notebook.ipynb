# <font color= #6B6B6B> <font size = 6> 1. Introduction </font>

<hr style="border:0.02in solid gray"> </hr>

This project aims to clone the operations of different recognized value investors in order to evaluate their performance. This can be achieved since the information of the quarterly operations of investors such as Warren Buffett, Charlie Munger or David Tepper are disclosed to the public through the $13$-F filings. However, it should be clarified that such operations, which only include the purchase and sale of shares, will be carried out with a delay of $45$ days, since this is the period that must pass from the end of any quarter for the information to be revealed. The ultimate goal of the project is to answer the following question: **can good returns be obtained by cloning the operations of superinvestors even when the information arrives with a certain delay?**

*NOTE: the necessary information is obtained from DATORAMA.* 

![](https://www.azquotes.com/picture-quotes/quote-i-believe-in-the-discipline-of-mastering-the-best-that-other-people-have-ever-figured-charlie-munger-59-5-0528.jpg)

## Assumptions

+ All the information regarding the operations of the investors arrives with a delay of $45$ days after the end of each quarter, consequently the purchase/sale of shares is executed with a delay of that amount of time. 
+ There is no initial capital, since it is intended to quantify the returns that would be obtained if the superinvestor's operations were cloned. 
+ Shares are only purchased when the superinvestor allocates capital for the first time (no further capital is added to such positions). Something similar happens with sales, since the position in the assets is only sold when the superinvestor gets rid of his complete stake (positions are not partially reduced either). 
+ Transaction costs are not considered.

## <font color= #6B6B6B> 1.2 Li Lu's Shameless Cloning </font>

For each cloned operation the following is displayed: 

+ **Ticker**. It is the stock's ticker symbol. 
+ **Timestamp (Buy)**. It is the date on which the purchase operation was carried out (considering the information delay).
+ **Buy Price**. Is the purchase price.
+ **Timestamp (Sell)**. It is the date on which the sale operation was carried out (considering the information delay).
+ **Sell Price**. Is the sale price.
+ **Effective Return %**. It is the total return obtained by the operation (in percentage).
+ **CAGR %**. It is the annualized return on investment (in percentage).
